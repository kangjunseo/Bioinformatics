```python
import pandas as pd
from dacon_submit_api import dacon_submit_api 
```

```python
dir_path = "/content/gdrive/MyDrive/open/"
test_df = pd.read_csv(dir_path + "test.csv")
train_df = pd.read_csv(dir_path + "train.csv")
```

```python
train_df
```

```python
train_df.drop(labels=['father','mother','gender'], axis=1, inplace=True)
train_df.head()
```

```python
from sklearn import svm
clf = svm.SVC(C=10, decision_function_shape='ovo')
embed = {"A A":0, "A G":1, "A C":2,
         "G A":3, "G G":4, "G C":5,
         "C A":6, "C G":7, "C C":8}
```

```python
X = train_df.drop(labels=['id', 'class'], axis=1)
X.replace(embed, inplace=True)
Y = train_df['class']
Y.replace({'A':0, 'B':2, 'C':1}, inplace=True)
clf.fit(X, Y)
```

```python
X_test = test_df.drop(labels=['id','father','mother','gender'], axis=1)
X_test.replace(embed, inplace=True)
X_test.head()
```

```python
Y_test = clf.predict(X_test)
```

```python
len(Y_test)
```

```python
y = pd.Series(Y_test).replace({0:'A', 2:'B', 1:'C'})
y.name = 'class'
ans = pd.concat([test_df['id'], y], axis=1)
ans
```

```python
print(list(ans['class']).count('A'))
print(list(ans['class']).count('B'))
print(list(ans['class']).count('C'))
```

```python
ans.to_csv('/content/out.csv', index=False)
```
